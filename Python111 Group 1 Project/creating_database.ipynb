```python
import pandas as pd
```

```python
df = pd.read_csv("onlinefoods.csv")
```

```python
df.head()
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('onlinefoods.db')
```

```python
# Replace 'your_table_name' with the name you want to give to your table
df.to_sql('foods', conn, if_exists='replace', index=False)
```

```python
cousor = conn.cursor()
cousor.execute("select * from foods").fetchall()
```

```python
conn.commit()
conn.close()
```
